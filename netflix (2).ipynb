```
netflix_shows=df[df['type']=='TV Show']
df_date= netflix_shows[['date_added']].dropna()
df_date['year'] = df_date['date_added'].apply(lambda x : x.split(', ')[-1])
df_date['month'] = df_date['date_added'].apply(lambda x : x.lstrip().split(' ')[0])
df_date
```

```

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'][::-1]


df_resultat = df_date.groupby('year')
df_resultat =df_resultat ['month'].value_counts().unstack().fillna(0)[month_order].T
df_resultat 

```

```
plt.figure(figsize=(10, 7), dpi=200)
plt.pcolor(df_resultat, cmap='Reds', edgecolors='white', linewidths=2) # heatmap
plt.xticks(np.arange(0.5, len(df_resultat.columns), 1), df_resultat.columns, fontsize=7, fontfamily='serif')
plt.yticks(np.arange(0.5, len(df_resultat.index), 1), df_resultat.index, fontsize=7, fontfamily='serif')

plt.title('Netflix Contents Update', fontsize=12, fontfamily='calibri', fontweight='bold', position=(0.20, 1.0+0.02))
cbar = plt.colorbar()

cbar.ax.tick_params(labelsize=8) 
cbar.ax.minorticks_on()
plt.show()
```

```
netflix_movies=df[df['type']=='Movie']
k=dict(netflix_movies['listed_in'].value_counts())


m=list(k.keys())
c=[]
for i in range(len(m)):
    ele=list(m[i].split(','))
    for item in ele:
        c.append(item)

list_final=c
out = []
for x in list_final:
    x=x.strip()
    if x not in out:
        out.append(x)


     
```

```
out
h=[0]*len(out)
f=dict(zip(out,h))

for ele in out:
    for item in k.keys():
        for j in item.split(','):
            if ele in  j.strip():
                f[ele]+=k[item]
f


```

```
f={k: v for k, v in sorted(f.items(), key=lambda item: item[1] ,reverse=True)}
f
```

```
fig, ax = plt.subplots()
fig = plt.figure(figsize = (10, 10))
x=list(f.keys())
y=list(f.values())
ax.vlines(x, ymin=0, ymax=y, color='green')
ax.plot(x,y, "o", color='red')
ax.set_xticklabels(x, rotation = 90)
ax.set_ylabel("nombre de films")

ax.set_title("types")
```
